Authorship attribution for Sanskrit philosophical texts: a small-scale statistical analysis

## Project outline

Question of authorship attribution are notoriously controversial in the case of ancient Indian religious and philosophical texts. Even though attempts to address attribution questions by statistical methods have a relatively long history, they have so far only rarely been undertaken in the study of Sanskrit texts. The aim of this project is to test the feasibility of this approach by looking at a relatively small set of texts. Our aim is to find out whether statistical methods can help in any way in settling two specific open questions about the authorship of some ancient Indian philosophical works.
Nāgārjuna (1st to 2nd century CE) is one of the greatest philosophical authors of ancient India. Unfortunately, many questions about Nāgārjuna remain unclear, including which of the many works attributed to him (the Tibetan canon attributes 116 different texts of very diverse content and quality to him) were in fact authored by Nāgārjuna.

# Objectives

Our aim is to find out whether the resulting analysis will allow us to make any progress in answering the following two questions:

1. It is sometimes questioned whether the final two chapters 26 and 27 of the Mūlamadhyamakakārikā are indeed part of the original work, or whether they are additions by a later hand.[3] Does a statistical comparison of chapters 1-25 with 26-27 provide confirming or disconfirming evidence?

1. That Nāgārjuna authored the Mūlamadhyamakakārikā is uncontroversial. Does the statistical comparison of other Sanskrit works ascribed to him allow us to differentiate between them in terms of similarity to the Mūlamadhyamakakārikā? In particular, would all of these texts be more similar to the Mūlamadhyamakakārikā than a contrasting text (composed around the same time and dealing with a similar subject-matter), Āryadeva’s Catuḥ śitaka?

[3] “Some modern scholars […] claim that the remaining two chapters are later additions and not the work of Nāgārjuna. In support of this claim they point out that the earliest of the existing commentaries, the Akutobhayā, might seem to have ended at this point. What are presented, in currently available editions of this commentary, as its last two chapters (i.e., commentary on chapters 26–27) are for the most part just the verses themselves, with no elucidatory comments. It might also be said in particular that chapter 26 presents no distinctively Madhyamaka views.” Mark Siderits, Shōryū Katsura: Nāgārjuna's Middle Way : the Mūlamadhyamakākarikā, Wisdom, Boston, 2013: 305. http://gretil.sub.uni-goettingen.de/gretil/1_sanskr/6_sastra/3_phil/buddh/naglok_u.htm


# Methods

We will run a statistical analysis of all of these texts, investigating in particular:

## Term Frequency \* Inverse Document Frequency (TF*IDF) 

TF*IDF is a way of expressing how important a word is to a document in a corpus: each word is marked against the overall corpus and the frequency value is offset by the frequency of the word in the corpus giving rare words more value. TF*IDF can produce a comparative map of themes, which can sever to highlight how different texts (and authors potentially) emphasized different features of the texts.

## Collocation Analysis

Examining which words appear near which other words in sentences is known as Collocation Analysis and could indicate author preferences in style of expression.

## Semantic Vector Space Mapping

Semantic Vector Space Mapping is a way of reducing all words in a corpus and the relationships between them to a set of vectors in a vector space, which does away with the need for semantically similar words to be near each other spatially in the text and affords us to have a mathematical representation of the semantic content of a document within a corpus.

## Minimum Description Length (MDL)

Finally MDL is a formalization of Occam’s Razor in which the best hypothesis for a given set of data is the one that leads to the best compression rate. Comparing compression rates across texts we can gain better understanding of the similarities and dissimilarities between posited authors.
