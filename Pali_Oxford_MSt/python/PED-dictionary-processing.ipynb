```python
import os
fileids = os.listdir('../PED')
ped = open("ped.xml","w")
for i in fileids:
    f = open('../PED/' + i, "r")
    raw = f.read()
    ped.write(raw)
    f.close()
ped.close()
```

```python
from bs4 import BeautifulSoup
import re
f = open('ped.xml')
raw = f.read()
f.close()
regexp = r"<d>(?: ){,3}&lt;b&gt;(.+?)&lt;/b&gt;"

vocab = re.findall(regexp, raw)
print(len(vocab))
# should be 16,282 "<d>"
#      7 with 1 space  "<d> &lt;b&gt;"
#(16,274 with 2 spaces "<d>  &lt;b&gt;")
#      1 with 3 spaces "<d>   &lt;b&gt;"
vocab[:100]
```

```python
s = "Anka&lt;sup&gt;1&lt;/sup&gt;"
s.find("&")
s[:4]
```

```python
def cut(word):
    n = word.find("&")
    return word[:n]

cut(s)
```

```python
vocab = [cut(w) for w in vocab]
vocab[:100]
```

```python
veq = [w for w in vocab if "=" in w]
print(len(veq))
veq
```

```python
vspace = [w for w in vocab if " " in w]
print(len(vspace))
vspace
```

```python
def strip(word):
    reg = re.findall(r"\w+", word)
    if len(reg) > 0:
        return reg[0]
    return word
```

```python
vocab2 = [strip(w).lower() for w in vocab]
vocab2[:100]
len(set(vocab2))
```

```python
sorted(set(vocab2))
```

```python
wordlist = sorted(set(vocab2))[1:]
f = open("PED-wordlist.txt", "w")
for w in wordlist:
    f.write(w + "\n")
f.close()
```

# How many headwords are there in the PED?

A discussion

With 16282 entries in the PED we face a few complications.

1. Equal sign in the headwords. There are a total of 36 such instances. We decided to ignore them but it would potentially add 36 to our count 16,282 + 36 = 16,318 (???)
1. Secondary, tertiary etc. meanings of a word are given as their own entries. Thus "A" appears 4 times. We decide to remove duplicates and count identical surface forms only once. Thus only one "A". This will count homonyms as one single headword. This gives us 15,014 headwords. (The first was an empty string, so it's removed)
1. Prefixes are listed as their own entries such as `a`, `abbh`, `ant` etc. These probably do not carry meaning by themselves. It might be better in the future to reduce the dictionary to headwords which carry a full semantic meaning on their own, that is: nouns, verbs, adjectives, adverbs, pronouns, and indeclinables only.
